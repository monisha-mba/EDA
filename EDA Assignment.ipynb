```python
def infer_Outlet_size(row):
    if pd.isnull(row['Outlet_Size']):
        if row['Item_Type'] in ['Fruits and Vegetables']:
            return 'Medium'
        elif row['Item_Type'] in ['Household']:
            return 'High'
        elif row['Outlet_Type'] == 'Grocery Store':
            return 'Small'
        elif row['Outlet_Type'] == 'Supermarket Type1':
            return 'Medium'
        else:
            return 'Small'
    else:
        return row['Outlet_Size']
    
```

```python
df['Outlet_Size'] = df.apply(infer_Outlet_size, axis=1)
```
